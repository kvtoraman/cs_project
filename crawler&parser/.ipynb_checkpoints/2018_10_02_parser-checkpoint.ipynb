```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from konlpy.tag import Kkma
from konlpy.utils import pprint
import json
from pathlib import Path
kkma = Kkma()
```

```python
mp = {}
def parse(data):
    nouns = kkma.pos(data["review_text"])
    for n in nouns:
        if n not in mp:
            mp[n] = 0
        mp[n] = mp[n] + 1
            
    return True
```

```python
REVIEWS = "../reviews_in_one_txt/all_reviews"
f = open("" + REVIEWS + ".txt","r",encoding="utf-8").readlines()
for line in f[:5]:
    course = json.loads(line)
    parse(course)
    review_text = course["review_text"].strip()
    
```

```python
folderpath = Path("reviews/")

START_ID = 2880
END_ID = 10000
TOTAL_REVIEWS = 5
current_reviews = 0
current_id = START_ID
while current_reviews < TOTAL_REVIEWS and current_id < END_ID: # try different IDs
    result = parse(current_id)
    if result:
        current_reviews += 1
    current_id += 1
```
