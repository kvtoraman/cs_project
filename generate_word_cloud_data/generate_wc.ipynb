```python
import json
import numpy as np
```

```python
REVIEWS = "cs_reviews"

f = open("../reviews_in_one_txt/" + REVIEWS + ".txt","r",encoding="utf-8").readlines()
g = open("../reviews_in_one_txt/" + REVIEWS + "_nouns.txt","r",encoding="utf-8").readlines()
```

```python
len(f)
```

```python
len(g)
```

```python
mp = {}
lls = []
index = {}
rev_index = {}
CNT = 0
for i in range(len(f)):
    course = json.loads(f[i])
    nouns = g[i].split()
    if course["course_code"] not in index:
        index[course["course_code"]] = CNT
        rev_index[CNT] = course["course_code"]
        CNT += 1
        lls.append([])
    for n in nouns:
        lls[index[course["course_code"]]].append(n)
        
    if course["course_code"] not in mp:
        mp[course["course_code"]] = {}
    for w in nouns:
        if w not in mp[course["course_code"]]:
            mp[course["course_code"]][w] = 0
        mp[course["course_code"]][w] += 1
```

```python
js = {}
for k,v in mp.items():
#     print(k,v)
    js[k] = [[a,b] for a,b in v.items()]
    
```

```python
with open("word_cloud_data.json","w",encoding="utf-8") as out:
    json.dump(js,out,ensure_ascii=False)
```
