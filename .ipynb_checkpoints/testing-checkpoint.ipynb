```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

```python
def parse(url_id):
    if url_id % 100 == 0:
        print(url_id)
    target_url = "https://otl.kaist.ac.kr/review/result/comment/" + str(url_id) + "/"
    target_request = Request(target_url)
    response = urlopen(target_request)
    response_text = response.read()
    soup = BeautifulSoup(response_text, "lxml")
    some_str = soup.find(id="contents").div.div.text
    if "없어용" in some_str:
        return
    course_code_name = soup.select("#" + str(url_id) + " > h4")[0].text
    print(course_code_name)
    prof_name_year_semester = soup.select("#" + str(url_id) + " > h4 > small")[0].text
    print(prof_name_year_semester)
    content = soup.select("#" + str(url_id) + " > p")
    print(content)
    letter_eval = soup.select("#contents > div > div > div > div > div > div:nth-of-type(5) > span > div > div")
    for letter in letter_eval:
        print(letter.text.strip().split("&")[0], letter.span.text)
    
START_ID = 6000
TOTAL_TRY = 10
for i in range(START_ID, START_ID + TOTAL_TRY): # try different IDs
    parse(i) 
```
