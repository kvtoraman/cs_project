```python
from konlpy.tag import Okt
import numpy as np
```

```python
okt = Okt()
word = "수학을"
```

```python
pos_list = okt.pos(word, norm=True)
print(pos_list)
word_vector = np.sum([pos_vectors.word_vec(str(pos).replace(" ", "")) for pos in pos_list], axis=0)
```
