```python
mypath = "../reviews"
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
import glob
all_review_paths = glob.glob(mypath + "/*.txt")
```

```python
onlyfiles
```

```python
college_years = ["freshman","sophomore","junior","senior"]
allowed_majors = ["CS"]
def get_year(course_code):
    if int(course_code[-3]) - 1 < 4:
        return college_years[int(course_code[-3]) - 1]
    else:
        return "graduate"
    
sentiment_mp = {True: "pos", False: "neg"}
subject_mp = {"강의":0,"널널":1,"성적":2}
rev_subject_mp = {0:"강의", 1:"널널", 2:"성적"}
    
def flare_from_json(course, is_pos, category):
    (course_code,course_name) = tuple(course["course_code_name"].split(":",1))
    course_code = course_code.strip()
    course_name = course_name.strip()
#     print("="+course_code+"=")
#     print(course_name)
    
    result = ""
    if course_code[:2] not in allowed_majors:
        return ""
    result += get_year(course_code) + "-"
    result += course_code + "-"
#     result += course["prof_name"] + "-"
    result += sentiment_mp[is_pos] + "-"
    result += rev_subject_mp[category]
#     result += course["year"] +course["semester"]
    
    
    return result

count_reviews = {}
for path in all_review_paths:
#     print(path)
    f = open(path,"r",encoding="utf-8")
    course = json.load(f)
    for sentiment in [True, False]:
        for category in [0,1,2]:
            row = flare_from_json(course,sentiment, category)
            if row == "":
                continue
            if row not in count_reviews:
                count_reviews[row] = 0
            count_reviews[row] += 1
```

```python
with open("csonly_reviews_withsentiment_in_flare.csv","w",encoding="utf-8") as out:
    for k,v in count_reviews.items():
        out.write(k+','+str(v)+"\n")
```
