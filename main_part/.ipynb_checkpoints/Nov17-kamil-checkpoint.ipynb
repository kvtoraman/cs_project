```python
import json
import numpy as np
from konlpy.tag import Kkma
from konlpy.utils import pprint
```

```python
def parse_long_str(s):
    s = s.split(',')
    s = [x.strip() for x in s]
    return s
```

```python
good_lecture = parse_long_str("사랑, 천사, 친절, 유명, 짱, 참여, 최고, 만족, 유익, 열정적, 갓, 답변, 명강의," + 
                             """잘 하심,잘 하십니다, 활발, 남는 게 많,남는 것이 많, 힐링, 수업을 잘, 질문,관심, 따라가기, 도움이 되, 추천, 친절, 들어야, 감사, 배우,재밌, 재미있, 필기""")
bad_lecture = parse_long_str("집중, 혼자, 부족, 자습, 당황, 최악, 인생, 불편, 복잡, 불만, 악명, 비추, 난이도," +
                             """답답, 후회,졸리,졸려,아쉬움,독학, 불친절, 부족, 못하십니다, 허술, 비추, 배우는 게, 모르, 따라가""")
```

```python
#todo try weighted average of w2v
scores = {}
def add_to_dict(l,score):
    global scores
    print("Added " + str(len(l)))
    for i in range(len(l)):
        scores[l[i]] = score
```

```python
add_to_dict(good_lecture,1.0)
```

```python
add_to_dict(bad_lecture,-1.0)
```

```python
def calculate_score(s):
    result = 0
    for k,v in scores.items():
        if k in s:
            result += v
    return result
```

```python
REVIEWS = "cs_reviews"
```

```python
f = open("../reviews_in_one_txt/" + REVIEWS + ".txt","r",encoding="utf-8").readlines()
```

```python
scores_list = []
for i in range(len(f[:])):
#     line = f[-i-20]
    line = f[i]
    course = json.loads(line)
#     print(course["review_text"])
#     print(calculate_score(course["review_text"]))
    scores_list.append(calculate_score(course["review_text"]))
#     print(course["강의"])
#     print("------------")
```

```python
mp = {"A":5,"B":4,"C":3,"D":2,"F":1}
for i in range(len(f[:])):
    course = json.loads(line)
    if course["강의"] != "?":
        print(mp[course["강의"]],(scores_list[i]+4)*5/11)
    
    
```

```python
with open("cs_reviews_morphs.txt","w",encoding="utf-8") as out:
    for i in range(len(f)):
        if i%100==0:
            print(i)
        line = f[i]
        course = json.loads(line)
        try:
            p_list = kkma.morphs(course["review_text"])
        except:
            p_list = []
            concat = " "
#         print(p_list)
#         concat = ["(%s,%s)"%(p[0],p[1]) for p in p_list]

        if len(p_list) > 0:
            concat = " ".join(p_list)  
        else:
            concat = " "

        out.write(concat + "\n")
    
```
